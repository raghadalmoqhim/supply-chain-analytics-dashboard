```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data = pd.read_csv("supply_chain_data.csv")
```

```python
print(data.head())
```

```python
print(data.describe())
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data.info()
```

```python
data.rename(columns={
    'Lead times': 'Inventory Lead Time',
    'Lead time': 'Supplier Lead Time',
    'Manufacturing lead time': 'Manufacturing Lead Time'
}, inplace=True)
```

```python
data.groupby("Product type")["Revenue generated"].sum().sort_values(ascending=False)
```

```python
data.groupby("Supplier name")["Defect rates"].mean().sort_values()
```

```python
data[["Shipping times", "Shipping costs"]].corr()
```

```python
data[data["Stock levels"] == 0]
```

```python
data.columns = data.columns.str.lower().str.replace(" ", "_")
```

```python
data[["shipping_costs", "manufacturing_costs", "costs"]].describe()
```

```python
data[["shipping_costs", "manufacturing_costs", "costs"]].head(10)
```

```python
data[["shipping_costs", "manufacturing_costs", "costs"]].corr()
```

```python
data.groupby("transportation_modes")["costs"].mean().sort_values()
```

```python
data.groupby(["transportation_modes", "routes"])["costs"].mean()
```

```python
data.groupby("transportation_modes")["shipping_times"].mean()
```

```python
data["cost_efficiency"] = data["shipping_times"] / data["costs"]
```

```python
data[["shipping_times", "costs", "cost_efficiency"]].head()
```

```python
data.groupby("transportation_modes")["cost_efficiency"].mean().sort_values()
```

```python
data.to_csv("clean_supply_chain.csv", index=False)
```

```python

engine = create_engine("mysql+pymysql://PASS@localhost/supply_chain_db")
data.to_sql("supply_chain", con=engine, if_exists="replace", index=False)

print("Data uploaded successfully!")
```
